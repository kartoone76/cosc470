```python
# PART 1 - single neuron network with different activation functions
%cd "/workspaces/cosc470/nn"
#%cd "~/cosc470/nn"
import importlib
import network2
import numpy as np
importlib.reload(network2)
net = network2.Network([1, 1], cost=network2.QuadraticCost)  
training_data = [np.array(([i%2], [(i+1)%2])) for i in range(100)]
test_data = [
    (np.array([[i % 2]], dtype=np.float32), np.int64((i + 1) % 2))
    for i in range(100)
]
net.weights = [np.array([[0.6]])]
net.biases = [np.array([[0.9]])]
print(net.weights)
print(net.biases)

# RUN YOUR EXPERIMENT HERE
(eval_cost_quad_good, _, _, _) = net.SGD(training_data, 300, 10, 0.02, evaluation_data=test_data, monitor_evaluation_cost=True)

net.weights = [np.array([[2.0]])]
net.biases = [np.array([[2.0]])]

(eval_cost_quad_bad, _, _, _) = net.SGD(training_data, 300, 10, 0.02, evaluation_data=test_data, monitor_evaluation_cost=True)

```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 6))

ax[0].plot(eval_cost_quad_good, label='Quadratic Cost Good Init')
ax[0].set_title('Quadratic Cost Good Init')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Cost')

ax[1].plot(eval_cost_quad_bad, label='Quadratic Cost Bad Init', color='orange')
ax[1].set_title('Quadratic Cost Bad Init')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Cost')
plt.tight_layout()
plt.show()
```

```python
# PART 1 - single neuron network with different activation functions
%cd "/workspaces/cosc470/nn"
#%cd "~/cosc470/nn"
import importlib
import network2
import numpy as np
importlib.reload(network2)
net = network2.Network([1, 1], cost=network2.CrossEntropyCost)  
training_data = [np.array(([i%2], [(i+1)%2])) for i in range(100)]
test_data = [
    (np.array([[i % 2]], dtype=np.float32), np.int64((i + 1) % 2))
    for i in range(100)
]
net.weights = [np.array([[0.6]])]
net.biases = [np.array([[0.9]])]
print(net.weights)
print(net.biases)

# RUN YOUR EXPERIMENT HERE
(eval_cost_cross_good, _, _, _) = net.SGD(training_data, 20, 10, 0.05, evaluation_data=test_data, monitor_evaluation_cost=True)

net.weights = [np.array([[2.0]])]
net.biases = [np.array([[2.0]])]

(eval_cost_cross_bad, _, _, _) = net.SGD(training_data, 20, 10, 0.05, evaluation_data=test_data, monitor_evaluation_cost=True)

```
